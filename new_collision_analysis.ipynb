```python
import matplotlib.pyplot as plt
import matplotlib as mlt
import numpy as np
import scipy
from scipy.interpolate import griddata
```

Below is the dump file analysis before the collision between two 1 jupiter mass polytropes of 300k particles at a separation distance of 55 jupiter radii with interpolation

```python
data = np.loadtxt('relax_00000.ascii')
data4 = np.loadtxt('collision_radialvelocity_00004.ascii')
data5 = np.loadtxt('collision_radialvelocity_00005.ascii')
data6 = np.loadtxt('collision_radialvelocity_00006.ascii')
data7 = np.loadtxt('collision_radialvelocity_00007.ascii')
data8 = np.loadtxt('collision_radialvelocity_00008.ascii')
data29 = np.loadtxt('collision_radialvelocity_00029.ascii')
data30 = np.loadtxt('collision_radialvelocity_00030.ascii')

radius_jupiter = np.sqrt(data[:,0]**2 + data[:,1]**2 + data[:,2]**2)
density = data[:,5]
temperature = data[:,6]

radius_jupiter4 = np.sqrt(data4[:,0]**2 + data4[:,1]**2 + data4[:,2]**2)
density4 = data4[:,5]
temperature4 = data4[:,6]

radius_jupiter5 = np.sqrt(data5[:,0]**2 + data5[:,1]**2 + data5[:,2]**2)
density5 = data5[:,5]
temperature5 = data5[:,6]
velocity5 = np.sqrt((data5[:,7] * 1000)**2 + (data5[:,8] * 1000)**2 + (data5[:,9] * 1000)**2)
momentum5 =  6.32910013698630012E+21 * velocity5

radius_jupiter6 = np.sqrt(data6[:,0]**2 + data6[:,1]**2 + data6[:,2]**2)
density6 = data6[:,5]
temperature6 = data6[:,6]
velocity6 = np.sqrt((data6[:,7] * 1000)**2 + (data6[:,8] * 1000)**2 + (data6[:,9] * 1000)**2)
momentum6 =  6.32910013698630012E+21 * velocity6

radius_jupiter7 = np.sqrt(data7[:,0]**2 + data7[:,1]**2 + data7[:,2]**2)
density7 = data7[:,5]
temperature7 = data7[:,6]
velocity7 = np.sqrt((data7[:,7] * 1000)**2 + (data7[:,8] * 1000)**2 + (data7[:,9] * 1000)**2)
momentum7 =  6.32910013698630012E+21 * velocity7

radius_jupiter8 = np.sqrt(data8[:,0]**2 + data8[:,1]**2 + data8[:,2]**2)
density8 = data8[:,5]
temperature8 = data8[:,6]
velocity8 = np.sqrt((data8[:,7] * 1000)**2 + (data8[:,8] * 1000)**2 + (data8[:,9] * 1000)**2)
momentum8 =  6.32910013698630012E+21 * velocity8

radius_jupiter29 = np.sqrt(data29[:,0]**2 + data29[:,1]**2 + data29[:,2]**2)
density29 = data29[:,5]
temperature29 = data29[:,6]
velocity29 = np.sqrt((data29[:,7] * 1000)**2 + (data29[:,8] * 1000)**2 + (data29[:,9] * 1000)**2)
momentum29 =  6.32910013698630012E+21 * velocity29

radius_jupiter30 = np.sqrt(data30[:,0]**2 + data30[:,1]**2 + data30[:,2]**2)
density30 = data30[:,5]
temperature30 = data30[:,6]
velocity30 = np.sqrt((data30[:,7] * 1000)**2 + (data30[:,8] * 1000)**2 + (data30[:,9] * 1000)**2)
momentum30 =  6.32910013698630012E+21 * velocity30
```

```python
radius = np.linspace(0,6,750)
density_interp = griddata(radius_jupiter, density, radius, method='nearest')
temperature_interp = griddata(radius_jupiter, temperature, radius, method='nearest')

density_interp4 = griddata(radius_jupiter4, density4, radius, method='nearest')
temperature_interp4 = griddata(radius_jupiter4, temperature4, radius, method='nearest')

density_interp5 = griddata(radius_jupiter5, density5, radius, method='nearest')
temperature_interp5 = griddata(radius_jupiter5, temperature5, radius, method='nearest')
momentum_interpolated5 = griddata(radius_jupiter5, momentum5, radius, method="nearest")

density_interp6 = griddata(radius_jupiter6, density6, radius, method='nearest')
temperature_interp6 = griddata(radius_jupiter6, temperature6, radius, method='nearest')
momentum_interpolated6 = griddata(radius_jupiter6, momentum6, radius, method="nearest")

density_interp7 = griddata(radius_jupiter7, density7, radius, method='nearest')
temperature_interp7 = griddata(radius_jupiter7, temperature7, radius, method='nearest')
momentum_interpolated7 = griddata(radius_jupiter7, momentum7, radius, method="nearest")

density_interp8 = griddata(radius_jupiter8, density8, radius, method='nearest')
temperature_interp8 = griddata(radius_jupiter8, temperature8, radius, method='nearest')
momentum_interpolated8 = griddata(radius_jupiter8, momentum8, radius, method="nearest")

density_interp29 = griddata(radius_jupiter29, density29, radius, method='nearest')
temperature_interp29 = griddata(radius_jupiter29, temperature29, radius, method='nearest')
momentum_interpolated29 = griddata(radius_jupiter29, momentum29, radius, method="nearest")

density_interp30 = griddata(radius_jupiter30, density30, radius, method='nearest')
temperature_interp30 = griddata(radius_jupiter30, temperature30, radius, method='nearest')
momentum_interpolated30 = griddata(radius_jupiter30, momentum30, radius, method="nearest")
```

```python
fig = plt.figure(figsize=(18,7.5))

plt.plot(radius,density_interp, lw=1.0, c='black', label='Density at Relaxed Dump 0')

plt.plot(radius, density_interp4, lw=1.0, c='purple', label='Density at Dump 4')

plt.plot(radius,density_interp5, lw=1.0, c='green', label='Density at Dump 5')

plt.plot(radius,density_interp6, lw=1.0, c='blue', label='Density at Dump 6')

plt.plot(radius,density_interp7, lw=1.0, c='red', label='Density at Dump 7')

plt.plot(radius,density_interp8, lw=1.0, c='orange', label='Density at Dump 8')

plt.plot(radius,density_interp29, lw=1.0, c='pink', label='Density at Dump 29')

plt.plot(radius,density_interp30, lw=1.0, c='brown', label='Density at Dump 30')

plt.legend()
plt.yscale('log')
plt.xlabel(r'$R \, [R_{\odot}]$', size=18)
plt.ylabel(r'log ($\rho$) $\, [g/cm^3]$', size=18)
plt.rc('xtick', labelsize=14) 
plt.rc('ytick', labelsize=14) 
```

```python
fig = plt.figure(figsize=(18,7.5))

plt.plot(radius,temperature_interp, lw=1.0, c='black', label='Temperature at Relaxed Dump 0')

plt.plot(radius,temperature_interp4, lw=1.0, c='purple', label='Temperature at Dump 4')

plt.plot(radius,temperature_interp5, lw=1.0, c='green', label='Temperature at Dump 5')

plt.plot(radius,temperature_interp6, lw=1.0, c='blue', label='Temperature at Dump 6')

plt.plot(radius,temperature_interp7, lw=1.0, c='red', label='Temperature at Dump 7')

plt.plot(radius,temperature_interp8, lw=1.0, c='orange', label='Temperature at Dump 8')

plt.plot(radius,temperature_interp29, lw=1.0, c='pink', label='Temperature at Dump 29')

plt.plot(radius,temperature_interp30, lw=1.0, c='brown', label='Temperature at Dump 30')

plt.legend()
plt.yscale('log')
plt.xlabel(r'$R \, [R_\odot]$', size=18)
plt.ylabel('log(T) [K]', size=18)
plt.rc('xtick', labelsize=14) 
plt.rc('ytick', labelsize=14) 
```
