```python
import matplotlib.pyplot as plt
import matplotlib as mlt
import numpy as np
import scipy
from scipy.interpolate import griddata
```

Below is the dump file analysis before the collision between two 1 jupiter mass polytropes of 200k particles at a separation distance of 55 jupiter radii with interpolation

```python
data3 = np.loadtxt('relax_00000.ascii')
data4 = np.loadtxt('relax_00167.ascii')
data5 = np.loadtxt('collision_radialvelocity_00005.ascii')
data6 = np.loadtxt('collision_radialvelocity_00006.ascii')
data7 = np.loadtxt('collision_radialvelocity_00007.ascii')
data11 = np.loadtxt('collision_radialvelocity_00011.ascii')
data12 = np.loadtxt('collision_radialvelocity_00012.ascii')

radius_jupiter3 = np.sqrt(data3[:,0]**2 + data3[:,1]**2 + data3[:,2]**2)
density3 = data3[:,5]
temperature3 = data3[:,6]

radius_jupiter4 = np.sqrt(data4[:,0]**2 + data4[:,1]**2 + data4[:,2]**2)
density4 = data4[:,5]
temperature4 = data4[:,6]

radius_jupiter5 = np.sqrt(data5[:,0]**2 + data5[:,1]**2 + data5[:,2]**2)
density5 = data5[:,5]
temperature5 = data5[:,6]

radius_jupiter6 = np.sqrt(data6[:,0]**2 + data6[:,1]**2 + data6[:,2]**2)
density6 = data6[:,5]
temperature6 = data6[:,6]

radius_jupiter7 = np.sqrt(data7[:,0]**2 + data7[:,1]**2 + data7[:,2]**2)
density7 = data7[:,5]
temperature7 = data7[:,6]

radius_jupiter11 = np.sqrt(data11[:,0]**2 + data11[:,1]**2 + data11[:,2]**2)
density11 = data11[:,5]
temperature11 = data11[:,6]

radius_jupiter12 = np.sqrt(data12[:,0]**2 + data12[:,1]**2 + data12[:,2]**2)
density12 = data12[:,5]
temperature12 = data12[:,6]
```

```python
radius = np.linspace(0,6,750)
denisty_interp3 = griddata(radius_jupiter3, density3, radius, method='nearest')
temperature_interp3 = griddata(radius_jupiter3, temperature3, radius, method='nearest')

denisty_interp4 = griddata(radius_jupiter4, density4, radius, method='nearest')
temperature_interp4 = griddata(radius_jupiter4, temperature4, radius, method='nearest')

denisty_interp5 = griddata(radius_jupiter5, density5, radius, method='nearest')
temperature_interp5 = griddata(radius_jupiter5, temperature5, radius, method='nearest')

denisty_interp6 = griddata(radius_jupiter6, density6, radius, method='nearest')
temperature_interp6 = griddata(radius_jupiter6, temperature6, radius, method='nearest')

denisty_interp7 = griddata(radius_jupiter7, density7, radius, method='nearest')
temperature_interp7 = griddata(radius_jupiter7, temperature7, radius, method='nearest')

denisty_interp11 = griddata(radius_jupiter11, density11, radius, method='nearest')
temperature_interp11 = griddata(radius_jupiter11, temperature11, radius, method='nearest')

denisty_interp12 = griddata(radius_jupiter12, density12, radius, method='nearest')
temperature_interp12 = griddata(radius_jupiter12, temperature12, radius, method='nearest')
```

```python
fig = plt.figure(figsize=(18,7.5))

plt.plot(radius,denisty_interp3, lw=1.0, c='black', label='Density at Relaxed Dump 0')

plt.plot(radius, denisty_interp4, lw=1.0, c='purple', label='Density at Relaxed Dump 167')

plt.plot(radius,denisty_interp5, lw=1.0, c='green', label='Density at Dump 5')

plt.plot(radius,denisty_interp6, lw=1.0, c='blue', label='Density at Dump 6')

plt.plot(radius,denisty_interp7, lw=1.0, c='red', label='Density at Dump 7')

plt.plot(radius,denisty_interp11, lw=1.0, c='orange', label='Density at Dump 11')

plt.plot(radius,denisty_interp12, lw=1.0, c='magenta', label='Density at Dump 12')

plt.legend()
plt.yscale('log')
plt.xlabel(r'$R \, [R_{\odot}]$', size=18)
plt.ylabel(r'log ($\rho$) $\, [g/cm^3]$', size=18)
plt.rc('xtick', labelsize=14) 
plt.rc('ytick', labelsize=14) 
```

```python
fig = plt.figure(figsize=(18,7.5))

plt.plot(radius,temperature_interp3, lw=1.0, c='black', label='Temperature at Relaxed Dump 0')

plt.plot(radius,temperature_interp4, lw=1.0, c='purple', label='Temperature at Relaxed Dump 167')

plt.plot(radius,temperature_interp5, lw=1.0, c='green', label='Temperature at Dump 5')

plt.plot(radius,temperature_interp6, lw=1.0, c='blue', label='Temperature at Dump 6')

plt.plot(radius,temperature_interp7, lw=1.0, c='red', label='Temperature at Dump 7')

plt.plot(radius,temperature_interp11, lw=1.0, c='orange', label='Temperature at Dump 11')

plt.plot(radius,temperature_interp12, lw=1.0, c='magenta', label='Temperature at Dump 12')

plt.legend()
plt.yscale('log')
plt.xlabel(r'$R \, [R_\odot]$', size=18)
plt.ylabel('log(T) [K]', size=18)
plt.rc('xtick', labelsize=14) 
plt.rc('ytick', labelsize=14) 
```
